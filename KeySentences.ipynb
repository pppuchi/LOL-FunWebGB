```python
#ç”¨æ­£åˆ™è¡¨è¾¾å¼æ‰¾å‡ºallytipså…³é”®å¥
import re
#æ‰“å¼€æ–‡ä»¶è¯»å–æ–‡ä»¶
with open(r'championFull.txt','r',encoding='utf-8') as r:
    r.text=r.read()

#ç¼–è¯‘æ­£åˆ™è¡¨è¾¾å¼
pattern_id=re.compile('id":\"(\w*?)\",\"key') #æå–idï¼ˆè‹±æ–‡åï¼‰
pattern_name=re.compile('\"key\":\"\d+\",\"name\":\"(.*?)\",\"title') #æå–åå­—ï¼ˆå­¦åï¼‰
pattern_title=re.compile('title":\"(.*?)\",\"image') #æå–ç§°å·ï¼ˆçœŸåï¼‰
pattern_allytips=re.compile('allytips":\[\"(.*?).\"],\"enemytips') #æå–å‹å†›æç¤º
pattern_enemytips=re.compile('enemytips":\[\"(.*?).\"],\"tags') #æå–æ•Œå†›æç¤º

#findallæŸ¥æ‰¾æŒ‡å®šå­—ç¬¦
i=re.findall(pattern_id,r.text)
n=re.findall(pattern_name,r.text)
t=re.findall(pattern_title,r.text)
a=re.findall(pattern_allytips,r.text)
e=re.findall(pattern_enemytips,r.text)

#clean list allytipsï¼Œgrouped by str
AllytipsClean=[]
for item in a:
    item=item.replace('\\','') #remove\
    item=item.replace('.","','<br>') #repalceã€‚","to <br> for website code
    item=item.strip()#delete space,unlike chinese, remove(space) will make the sentences one long word so use strip here
    AllytipsClean.append(item)

#clean list enemytipsï¼Œgrouped by str
EnemytipsClean=[]
for item in e:
    item=item.replace('\\','') #remove\
    item=item.replace('.","','<br>') #repalceã€‚","to <br> for website code
    item=item.strip()#delete space,unlike chinese, remove(space) will make the sentences one long word so use strip here
    EnemytipsClean.append(item)

#æ‰¹é‡åˆ¶ä½œç½‘é¡µä»£ç 
#èŽ·å–æ–‡ä»¶å¤¹å†…æ‰€æœ‰å›¾ç‰‡çš„åç§°
import os
SplashPath='splash'
SplashName=os.listdir(SplashPath)

for x in range(len(n)):
    para=f'<article class="thumb"><a href="splash/{SplashName[x]}" class="image"><img src="tiles/{SplashName[x]}" alt="" /></a><h2>{n[x]} {t[x]}</h2><p>{AllytipsClean[x]}</p><p>{EnemytipsClean[x]}</p></article>'
    print(para)
```
